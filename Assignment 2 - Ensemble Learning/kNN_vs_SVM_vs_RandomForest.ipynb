The exercise 2.3) are centered around the toy data set of classifying hand
written digits. The data set can be retrieved from load digits using sklearn. The data set
contains 1797 grey scale 8 × 8 images and corresponding labels (a number between 0 to 9 per
image). Although images usually are represented in two dimensions (height and width), with
load digits each image constitutes a single feature vector of 64 elements.
In order to obtain training and test data, we first have to split the data set between training
and test data. One can use the train test split API from python’s sklearn for this purpose. By
default, for train test split, we will keep 25% of the data as test set.

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.svm import SVC
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

# Read data and split into train and test

```python
digits = load_digits(return_X_y=True)

train, test, train_label, test_label = train_test_split(digits[0], digits[1])
```

## k-NN

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(train, train_label)
knn_prediction = neigh.predict(test)
knn_score = neigh.score(test, test_label)
knn_matrix = confusion_matrix(test_label, knn_prediction)
```

## SVM

```python
svm = SVC(kernel="linear")
svm.fit(train, train_label)
svm_predictions = svm.predict(test)
svm_score = svm.score(test, test_label)
svm_matrix = confusion_matrix(test_label, svm_predictions)
```

## Random Forest

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(train, train_label)
random_forest_predictions = random_forest.predict(test)
random_forest_score = random_forest.score(test, test_label)
random_forest_matrix = confusion_matrix(test_label, random_forest_predictions)
```

# k-NN Confusion matrix
Using KNeighborsClassifier from Scikit learn with n=3 neighbours.

```python
sn.set(font_scale=1.4)#for label size
sn.heatmap(knn_matrix, annot=True,annot_kws={"size": 16})

plt.show()
```

# SVM Confusion matrix
Using SVC from Scikit learn, with linear kernel.

```python
sn.set(font_scale=1.4)#for label size
sn.heatmap(svm_matrix, annot=True,annot_kws={"size": 16})# font size

plt.show()
```

# Random Forest matrix
Using RandomForestClassifier from Scikit learn with n=100 estimators.

```python
sn.set(font_scale=1.4)#for label size
sn.heatmap(random_forest_matrix, annot=True,annot_kws={"size": 16})# font size

plt.show()
```
